While Loops

```python

started = False

while True:
    command = input("> ").lower()
    if command == "start":
        if started:
            print(f'The car is already started')
        else:
            started = True
            print(f'The car started')

    elif command == "stop":
        if not started:
            print(f"The Car already stopped")
        else:
            started = False
            print(f"The Car stopped")
    elif command == 'help':
        print("""
              start - to start the car
              stop - to stop the car
              quit - to quit
              """)
    elif command == 'quit':
        break
    else:
        print(f'sorry i dont understand')
```
